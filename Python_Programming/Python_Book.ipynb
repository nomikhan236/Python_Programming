### Break Statement
if you have stop the loop even the condition becomes true you can use the break statement as follow:

```python
# print numbers upto 10 but break at 5
i = 0
while (i<=10):
    print(i)
    if (i == 5):
        break
    i += 1
```

### Continue Statement
If you have to stop at certein point in your code skip some values and then continue again you can use continue statement 

```python
# Example
# print numbers upto 10 but skip the value number 5
i = 0
while (i<=10):
        i += 1
        if (i == 5):
            continue
        print(i)
```
